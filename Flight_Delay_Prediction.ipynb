```python
def parse_crs_dep_time(row):
    time_str = str(row['CRS_DEP_TIME']).zfill(4)
    hour = int(time_str[:2])
    minute = int(time_str[2:])
    return pd.Timestamp(year=row['YEAR'], month=row['MONTH'], day=row['DAY_OF_MONTH'], hour=hour, minute=minute)

df_flights['dep_datetime'] = df_flights.apply(parse_crs_dep_time, axis=1)
```

```python
sort_df_flights = df_flights.sort_values('dep_datetime')
january_df_w['DATE'] = pd.to_datetime(january_df_w['DATE'])
j_df_w = january_df_w.rename(columns={'DATE': 'datetime'})
j_df_w = j_df_w.sort_values('datetime').reset_index(drop=True)
```

```python
merged_df = pd.merge_asof(
    sort_df_flights.sort_values('dep_datetime'),
    j_df_w.sort_values('datetime'),
    left_on='dep_datetime',
    right_on='datetime',
    tolerance=pd.Timedelta('30 minutes'),
    direction='nearest'
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from category_encoders import TargetEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay
from xgboost import XGBClassifier


df_model = df_filtered.copy()
df_model['DELAY_CLASS'] = (df_model['DEP_DELAY_NEW'] >= 25).astype(int)


df_0 = df_model[df_model['DELAY_CLASS'] == 0]
df_1 = df_model[df_model['DELAY_CLASS'] == 1]
df_0_reduced = df_0.sample(frac=0.46, random_state=42)
df_model = pd.concat([df_0_reduced, df_1], ignore_index=True).sample(frac=1, random_state=42).reset_index(drop=True)


X = df_model.drop(columns=['DEP_DELAY_NEW', 'DELAY_CLASS', 'OP_CARRIER_AIRLINE_ID',
                           'TAIL_NUM', 'OP_CARRIER_FL_NUM', 'DEST_CITY_NAME', 'ROUTE','IS_WEDNESDAY'])
y = df_model['DELAY_CLASS']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)


cat_cols = ['OP_UNIQUE_CARRIER', 'DEST_STATE_ABR', 'VISIBILITY_CAT',
            'HUMIDITY_CAT', 'DIST_BIN', 'FLIGHT_DATE', 'DEP_TIME_BLOCK','ORIGIN','DEST']
te = TargetEncoder(cols=cat_cols)
X_train_enc = te.fit_transform(X_train, y_train)
X_test_enc = te.transform(X_test)


print("Sınıf Dağılımı (Train):")
print(y_train.value_counts(normalize=True))
print("\nSınıf Dağılımı (Test):")
print(y_test.value_counts(normalize=True))


best_model = XGBClassifier(
    n_estimators=400,
    max_depth=9,
    learning_rate=0.02,
    subsample=0.84334,
    colsample_bytree=0.9,
    gamma=0.7,
    min_child_weight=9,
    scale_pos_weight=1.8,
    random_state=42
)


best_model.fit(X_train_enc, y_train)

y_probs = best_model.predict_proba(X_test_enc)[:, 1]
y_pred1 = best_model.predict(X_test_enc)

print("\nClassification Report:")
print(classification_report(y_test, y_pred1))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred1))

print("\nROC AUC:", round(roc_auc_score(y_test, y_probs), 4))

RocCurveDisplay.from_predictions(y_test, y_probs)
plt.title("ROC Curve - XGBoost (Optimized)")
plt.show()
```
